```python
import pandas as pd
import matplotlib.pyplot as plt


url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación

# Ingreso total por cada tienda




```python
# Calculo de ingresos por tienda
ingreso1 = tienda1['Precio'].sum()
ingreso2 = tienda2['Precio'].sum()
ingreso3 = tienda3['Precio'].sum()
ingreso4 = tienda4['Precio'].sum()

# Muestro resultados
print(f"Ingreso total Tienda 1: {ingreso1:,.2f}")
print(f"Ingreso total Tienda 2: {ingreso2:,.2f}")
print(f"Ingreso total Tienda 3: {ingreso3:,.2f}")
print(f"Ingreso total Tienda 4: {ingreso4:,.2f}")
```

# 2. Ventas por categoría

```python
# Calculo de conteo de productos por categoría en cada tienda
categoria_tienda1 = tienda1.groupby('Categoría del Producto').size()
categoria_tienda2 = tienda2.groupby('Categoría del Producto').size()
categoria_tienda3 = tienda3.groupby('Categoría del Producto').size()
categoria_tienda4 = tienda4.groupby('Categoría del Producto').size()

# Muestro resultados
print("Categoría en Tienda 1:")
print(categoria_tienda1)
print("\nCategoría en Tienda 2:")
print(categoria_tienda2)
print("\nCategoría en Tienda 3:")
print(categoria_tienda3)
print("\nCategoría en Tienda 4:")
print(categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
# Calculo de calificaciones promedio por tienda
calificacion_promedio_tienda1 = tienda1['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

# Muestro resultados
print(f"Calificación promedio Tienda 1: {calificacion_promedio_tienda1:.2f}")
print(f"Calificación promedio Tienda 2: {calificacion_promedio_tienda2:.2f}")
print(f"Calificación promedio Tienda 3: {calificacion_promedio_tienda3:.2f}")
print(f"Calificación promedio Tienda 4: {calificacion_promedio_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
# Función para identificar productos más y menos vendidos
def identificar_productos_vendidos(df, nombre_tienda):
    ventas_por_producto = df['Producto'].value_counts()
    producto_mas_vendido = ventas_por_producto.idxmax()
    producto_menos_vendido = ventas_por_producto.idxmin()

    print(f"Tienda {nombre_tienda}:")
    print(f"Producto más vendido: {producto_mas_vendido} (Ventas: {ventas_por_producto.max()})")
    print(f"Producto menos vendido: {producto_menos_vendido} (Ventas: {ventas_por_producto.min()})\n")

# Identificar productos más y menos vendidos en cada tienda
identificar_productos_vendidos(tienda1, "1")
identificar_productos_vendidos(tienda2, "2")
identificar_productos_vendidos(tienda3, "3")
identificar_productos_vendidos(tienda4, "4")
```

# 5. Envío promedio por tienda

```python
# Calculo de costo de envío promedio por tienda
envio_promedio_tienda1 = tienda1['Costo de envío'].mean()
envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

# Muestro resultados
print(f"Costo de envío promedio Tienda 1: {envio_promedio_tienda1:.2f}")
print(f"Costo de envío promedio Tienda 2: {envio_promedio_tienda2:.2f}")
print(f"Costo de envío promedio Tienda 3: {envio_promedio_tienda3:.2f}")
print(f"Costo de envío promedio Tienda 4: {envio_promedio_tienda4:.2f}")
```

# 6. Graficos

```python
# Calculo de distribución de categorías de productos
distribucion_categorias = pd.concat([tienda1, tienda2, tienda3, tienda4])['Categoría del Producto'].value_counts()

# Gráfico de barras para ingresos totales por tienda
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso1, ingreso2, ingreso3, ingreso4]

plt.figure(figsize=(10, 6))
plt.bar(tiendas, ingresos, color=['blue', 'green', 'red', 'purple'])
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos Totales')
plt.show()
```

**Interpretacion:**

Ingresos Totales por Tienda

Se observa que la Tienda 1 supera por poco a las demas Tiendas, y la Tienda 4 es la menos ventas tiene.

*   Ingreso total Tienda 1: 1,150,880,400.00
*   Ingreso total Tienda 2: 1,116,343,500.00
*   Ingreso total Tienda 3: 1,098,019,600.00
*   Ingreso total Tienda 4: 1,038,375,700.00



```python
# Gráfico de pastel para distribución de categorías de productos
plt.figure(figsize=(8, 8))
plt.pie(distribucion_categorias, labels=distribucion_categorias.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Categorías de Productos')
plt.show()
```

**Interpretacion:**

Gráfico de pastel de distribución de categorías de productos

La categoría de "Muebles" es la más vendida, seguida de "Electrónicos" y "Juguetes".


```python
# Gráfico de líneas para calificaciones promedio por tienda
calificaciones = [calificacion_promedio_tienda1, calificacion_promedio_tienda2, calificacion_promedio_tienda3, calificacion_promedio_tienda4]

plt.figure(figsize=(10, 6))
plt.plot(tiendas, calificaciones, marker='o', linestyle='-', color='brown')
plt.title('Calificaciones Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.show()
```

**Interpretacion:**

Gráfico de líneas de calificaciones promedio por tienda

La Tienda 3 tiene la calificación promedio más alta, seguida de la Tienda 2, Tienda 4 y Tienda 1.

# 7. Extra

```python
# Función para graficar la distribución geográfica de las ventas
def graficar_distribucion_geografica(df, nombre_tienda):
    plt.figure(figsize=(10, 8))
    plt.scatter(df['lon'], df['lat'], alpha=0.5, s=10)
    plt.title(f'Distribución Geográfica de Ventas - {nombre_tienda}')
    plt.xlabel('Longitud')
    plt.ylabel('Latitud')
    plt.grid(True)
    plt.show()

# Graficar distribución geográfica para cada tienda
graficar_distribucion_geografica(tienda1, 'Tienda 1')
graficar_distribucion_geografica(tienda2, 'Tienda 2')
graficar_distribucion_geografica(tienda3, 'Tienda 3')
graficar_distribucion_geografica(tienda4, 'Tienda 4')
```
